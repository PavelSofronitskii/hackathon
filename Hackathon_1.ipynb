```python
import pandas as pd
```

```python
cargoes = pd.DataFrame()

columns_name=['filling_space_percent','density_percent','cargo_length', 'cargo_height', 'cargo_width', 'cargo_type', 'cargo_title',  'cargo_id', 'calculation_id']

for i in columns_name:
    cargoes.insert(0, i, pd.NA)

cargoes.loc[0] = pd.NA
```

```python
boxes = pd.DataFrame()
columns_name=['calculated_size_length', 'calculated_size_height', 'calculated_size_width','turnover','stacking','box_position_z','box_position_y','box_position_x' ,'size_length', 'size_height', 'size_width', 'cargo_id', 'box_title',  'box_id', 'calculation_id']

for i in columns_name:
    boxes.insert(0, i, pd.NA)

boxes.loc[0] = pd.NA
```

```python
name_of_jsons = pd.read_csv('/content/drive/MyDrive/names_of_jsons.csv')
```

```python
for i in name_of_jsons['NUM']:
    df = pd.read_json(f"/content/drive/MyDrive/ALGORITM/data_for_algoritm_{i}.json")
    
    new_row = pd.Series(data={'calculation_id':df['data']['calculation_id'], 
                              'cargo_id':df['data']['cargo_space']['id'], 
                              'cargo_title':df['data']['cargo_space']['title'],
                              'cargo_type':df['data']['cargo_space']['type'],
                              'cargo_width':df['data']['cargo_space']['loading_size']['width'],
                              'cargo_height':df['data']['cargo_space']['loading_size']['height'],
                              'cargo_length':df['data']['cargo_space']['loading_size']['length'],
                              'density_percent':df['data_result']['cargo_space']['calculation_info']['density_percent'],
                              'filling_space_percent':df['data_result']['cargo_space']['calculation_info']['filling_space_percent']})
    cargoes = cargoes.append(new_row, ignore_index=True)


    
    for i in range(len(df.data_result.boxes)):
        new_row = pd.Series(data={'calculation_id':df['data']['calculation_id'], 
                              'box_id':df['data_result']['boxes'][i]['id'], 
                              'box_title':df['data_result']['boxes'][i]['info']['title'],
                              'cargo_id':df['data_result']['boxes'][i]['cargo_id'],
                              'size_width':df['data_result']['boxes'][i]['size']['width'],
                              'size_height':df['data_result']['boxes'][i]['size']['height'],
                              'size_length':df['data_result']['boxes'][i]['size']['length'],
                              'box_position_x':df['data_result']['boxes'][i]['position']['x'],
                              'box_position_y':df['data_result']['boxes'][i]['position']['y'],
                              'box_position_z':df['data_result']['boxes'][i]['position']['z'],
                              'stacking':df['data_result']['boxes'][i]['stacking'],
                              'turnover':df['data_result']['boxes'][i]['turnover'],
                              'calculated_size_width':df['data_result']['boxes'][i]['calculated_size']['width'],
                              'calculated_size_height':df['data_result']['boxes'][i]['calculated_size']['height'],
                              'calculated_size_length':df['data_result']['boxes'][i]['calculated_size']['length']})
        boxes = boxes.append(new_row, ignore_index=True)
```
