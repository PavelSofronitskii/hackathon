```python
import pandas as pd
```

```python
boxes = pd.read_csv('/content/boxes.csv')
```

```python
boxes
```

```python
boxes  = boxes.drop(columns=['box_id', 'box_position_z','box_position_y','box_position_x' ,'cargo_id', 'box_title', 'Unnamed: 0', 'size_width', 'size_height', 'size_length'])
```

```python
boxes.loc[boxes['stacking'] == False, 'stacking'] = 0
boxes.loc[boxes['stacking'] == True, 'stacking'] = 1
boxes.loc[boxes['turnover'] == False, 'turnover'] = 0
boxes.loc[boxes['turnover'] == True, 'turnover'] = 1
```

```python
boxes
```

```python
cargoes = pd.read_csv('/content/cargoes.csv')
```

```python
cargoes['boxes'] = 0
```

```python
df = boxes.loc[1:2]
df = df.drop(index=1)
df = df.drop(index=2)
df
```

```python
for i in range(len(cargoes)):
    cargoes['boxes'][i] = df
```

```python
for i in range(len(cargoes)):
    for t in range(len(boxes)):
        if boxes['calculation_id'][t] == cargoes['calculation_id'][i]:
             cargoes['boxes'][i] = cargoes['boxes'][i].append(boxes.loc[t])
```

```python
cargoes  = cargoes.drop(columns=['Unnamed: 0', 'calculation_id', 'cargo_id', 'cargo_title', 'cargo_type'])
```

```python
cargoes['box_count'] = 0
```

```python
for i in range(len(cargoes)):
    cargoes['box_count'][i] = len(cargoes['boxes'][i])
```

```python
cargoes['box_unique'] = 0
```

```python
for i in range(len(cargoes)):
    cargoes['box_unique'][i] = cargoes['boxes'][i].value_counts().count()
```

```python
cargoes
```

```python
cargoes['boxes'][0]['calculated_size_width'].sum()
```

```python
cargoes.describe()
```

```python
cargoes['box_sum_size_width'] = 0
cargoes['box_mean_size_width'] = 0
cargoes['box_sum_size_height'] = 0
cargoes['box_mean_size_height'] = 0
cargoes['box_sum_size_length'] = 0
cargoes['box_mean_size_length'] = 0
```

```python
for i in range(len(cargoes)):
    cargoes['box_sum_size_width'][i] = cargoes['boxes'][i]['calculated_size_width'].sum()
    cargoes['box_sum_size_height'][i] = cargoes['boxes'][i]['calculated_size_height'].sum()
    cargoes['box_sum_size_length'][i] = cargoes['boxes'][i]['calculated_size_length'].sum()
    cargoes['box_mean_size_width'][i] = cargoes['boxes'][i]['calculated_size_width'].mean()
    cargoes['box_mean_size_height'][i] = cargoes['boxes'][i]['calculated_size_height'].mean()
    cargoes['box_mean_size_length'][i] = cargoes['boxes'][i]['calculated_size_length'].mean()
```
